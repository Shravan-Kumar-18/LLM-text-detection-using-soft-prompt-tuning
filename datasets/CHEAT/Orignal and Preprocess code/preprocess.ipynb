```python
df = pd.concat([human_df, ai_df], ignore_index=True)
```

```python
raw_keywords = df['keyword'].dropna().astype(str)
```

```python
unique_keywords = set()
```

```python
import re

for row in raw_keywords:
    # This regex looks for anything inside double quotes: "text" -> text
    matches = re.findall(r'"([^"]+)"', row)
    
    for match in matches:
        clean_keyword = match.strip().lower()
        if clean_keyword:
            unique_keywords.add(clean_keyword)
```

```python
print(f"Found {len(unique_keywords)} unique keywords. Saving to file...")
with open('unique_keywords.txt', 'w', encoding='utf-8') as f:
    for kw in sorted(unique_keywords):
        f.write(kw + '\n')
```

```python
def accurate_domain_classifier(keyword_string):
    # Handle empty or NaN values
    if pd.isna(keyword_string):
        return 'Unclassified'
    
    # Extract the individual keywords from the '"Word1","Word2"' format
    words = [w.strip().lower() for w in re.findall(r'"([^"]+)"', str(keyword_string))]
    
    # Check if ANY of the extracted words exist in our sets
    is_med = any(w in med_keywords for w in words)
    is_subj = any(w in subj_keywords for w in words)
    is_algo = any(w in algo_keywords for w in words)
    
    # Priority classification: M and S override A (to prevent IEEE CS bias)
    if is_med:
        return 'M'
    elif is_subj:
        return 'S'
    elif is_algo:
        return 'A'
    else:
        return 'Unclassified'
```

```python
print("Applying classification logic...")
df['domain'] = df['keyword'].apply(accurate_domain_classifier)
```

```python
def extract_perfect_samples(full_df, domain_code, sample_size=1000):
    subset = full_df[full_df['domain'] == domain_code]
    
    pos_count = len(subset[subset['label'] == 0])
    neg_count = len(subset[subset['label'] == 1])
    
    print(f"Domain {domain_code}: Found {pos_count} Human, {neg_count} AI abstracts.")
    
    if pos_count >= sample_size and neg_count >= sample_size:
        pos = subset[subset['label'] == 0].sample(sample_size, random_state=42)
        neg = subset[subset['label'] == 1].sample(sample_size, random_state=42)
        return pd.concat([pos, neg])
    else:
        print(f"WARNING: Shortfall in domain {domain_code}. Found {min(pos_count, neg_count)} usable pairs.")
        # If it falls short, it returns the maximum balanced dataset available
        min_size = min(pos_count, neg_count)
        pos = subset[subset['label'] == 0].sample(min_size, random_state=42)
        neg = subset[subset['label'] == 1].sample(min_size, random_state=42)
        return pd.concat([pos, neg])
```

```python
print("\n--- Final Extraction Results ---")
df_A = extract_perfect_samples(df, 'A')
df_M = extract_perfect_samples(df, 'M')
df_S = extract_perfect_samples(df, 'S')
```
