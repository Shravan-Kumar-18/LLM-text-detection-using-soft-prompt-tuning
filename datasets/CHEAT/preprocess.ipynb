```python
import pandas as pd
```

```python
def clean_and_shuffle_data(file_paths: list, random_seed: int = 42) -> dict:
    processed_datasets = {}
    
    for file_path in file_paths:
        df = pd.read_csv(file_path)
        df = df.drop(columns=['id'], errors='ignore')
        df = df.sample(frac=1, random_state=random_seed).reset_index(drop=True)
        
        processed_datasets[file_path] = df
        print(f"Processed {file_path}: Shuffled and dropped 'id'. Total rows: {len(df)}")
        
    return processed_datasets
```

```python
files = ['Algorithm.csv', 'Medicine.csv', 'Subject.csv']
datasets = clean_and_shuffle_data(files)
```

```python
datasets['Algorithm.csv'].to_csv('Algorithm.csv' , index = False)
```

```python
datasets['Medicine.csv'].to_csv('Medicine.csv' , index = False )
```

```python
datasets['Subject.csv'].to_csv('Subject.csv' , index = False)
```
