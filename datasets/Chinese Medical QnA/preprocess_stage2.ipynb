```python
import pandas as pd 
import os 
```

```python
def process_and_merge_domain(human_path: str, ai_path: str, domain: str) -> None:
    if not os.path.exists(human_path) or not os.path.exists(ai_path):
        print(f"Error: Missing files for domain '{domain}'.")
        return

    human_df = pd.read_csv(human_path, encoding='utf-8-sig')
    ai_df = pd.read_csv(ai_path, encoding='utf-8-sig')
    
    # Standardize answer columns
    rename_map = {'human_answer': 'answer', 'ai_answer': 'answer', 'gpt_answer': 'answer'}
    human_df.rename(columns=rename_map, inplace=True, errors='ignore')
    ai_df.rename(columns=rename_map, inplace=True, errors='ignore')
    
    # Assign labels (0 = Human, 1 = AI)
    human_df['label'] = 0
    ai_df['label'] = 1
    
    # Drop 'department'
    human_df.drop(columns=['department'], inplace=True, errors='ignore')
    ai_df.drop(columns=['department'], inplace=True, errors='ignore')
    
    # Concatenate - Pandas naturally aligns 'title', 'question', 'answer', and 'label'
    res = pd.concat([human_df, ai_df], ignore_index=True)
    
    # Slice exactly what we want and drop empty answers
    res = res[['title', 'question', 'answer', 'label']].dropna(subset=['answer'])
    
    # Shuffle
    res = res.sample(frac=1, random_state=42).reset_index(drop=True)
    
    # Export
    output_filename = f"{domain}_2000_AI_HUMAN.csv"
    res.to_csv(output_filename, index=False, encoding='utf-8-sig')
    print(f"Successfully generated {output_filename} with {len(res)} rows.")
```

```python
process_and_merge_domain('Human datasets/IM 1000 Human.csv' , 'Ai Datasets/IM 1000 AI.csv' , 'IM')
```
